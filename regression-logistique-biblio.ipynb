```python
import numpy as np  # Importe NumPy pour manipuler les tableaux et générer des valeurs aléatoires.
import matplotlib.pyplot as plt  # Importe matplotlib pour afficher les graphiques.
from sklearn.linear_model import LogisticRegression  # Importe le modèle de régression logistique de scikit-learn.
```

```python
np.random.seed(0)  # Fixe la graine aléatoire pour obtenir les mêmes données à chaque exécution.
X = (10 * np.random.rand(100, 1))  # Génère 100 valeurs aléatoires entre 0 et 10.
y = (X[:, 0] > 5).astype(int)  # Attribue la classe 1 si la valeur est supérieure à 5, sinon 0.

noise_idx = np.random.choice(100, 10, replace=False)  # Sélectionne 10 indices aléatoires pour ajouter du bruit.
y[noise_idx] = 1 - y[noise_idx]  # Inverse la classe sur ces indices pour rendre les données moins parfaites.

model = LogisticRegression()  # Crée une instance du modèle de régression logistique.
model.fit(X, y)  # Entraîne le modèle sur les données X et y.

X_test = np.linspace(X.min() - 1, X.max() + 1, 200).reshape(-1, 1)  # Génère des points réguliers pour tracer une courbe lisse.
y_prob = model.predict_proba(X_test)[:, 1]  # Calcule la probabilité d'appartenir à la classe 1 pour chaque point de test.

plt.figure(figsize=(10, 5))  # Crée une figure de taille 10 par 5.
plt.scatter(X, y, color='blue', label='Données observées')  # Affiche les points du jeu de données.
plt.plot(X_test, y_prob, color='orange', linewidth=2, label='Courbe logistique')  # Trace la courbe de probabilité prédite.
plt.axhline(y=0.5, color='red', linestyle='--', label='Seuil de décision')  # Trace la ligne horizontale correspondant au seuil de décision.
plt.title('Régression logistique')  # Ajoute un titre au graph.
plt.xlabel('Variable X')  # Ajoute un label à l'axe des abscisses.
plt.ylabel('Probabilité de la classe 1')  # Ajoute un label à l'axe des ordonnées.
plt.legend()  # Affiche la légende.
plt.grid(True, alpha=0.3)  # Ajoute une grille légère au graph.
plt.show()  # Affiche le graph final.
```
