```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
x_test, y_test = make_regression(n_samples=200, n_features=2, noise=10)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color='red', label='Données réelles')
plt.show()
```

```python
# Créer le modèle de regression linéaire
model = LinearRegression()
# Entraîner le modèle sur les données
model.fit(x_test, y_test)
# Faire des prédictions
y_prediction = model.predict(x_test)

# Afficher les coefficients et l'intercept du modèle
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Utiliser matplotlib pour visualiser les données et la ligne de régression en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color='red', label='Données réelles')
#Afficher les prédictions come un plan
ax.plot_trisurf(x_test[:, 0], x_test[:, 1], y_prediction, color='blue', alpha=0.5, label='Plan de régression')

ax.set_xlabel('1')
ax.set_ylabel('2')
ax.set_zlabel('Target')

ax.legend()
plt.show()
```
