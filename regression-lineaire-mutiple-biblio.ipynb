```python
import numpy as np  # Importe la bibliothèque NumPy et lui donne l'alias np pour manipuler des tableaux et faire des calculs numériques.
from sklearn.linear_model import LinearRegression  # Importe la classe LinearRegression de scikit-learn pour créer un modèle de régression linéaire.
from sklearn.datasets import make_regression  # Importe la fonction make_regression pour générer un jeu de données de régression artificiel.
import matplotlib.pyplot as plt  # Importe le module pyplot de matplotlib sous le nom plt pour créer des graphiques.
```

```python
np.random.seed(0)  # Fixe la graine aléatoire à 0 pour obtenir les mêmes données à chaque exécution.
x_test, y_test = make_regression(n_samples=200, n_features=2, noise=10)  # Génère 200 échantillons avec 2 variables explicatives et un bruit de 10.
fig = plt.figure() # Crée une nouvelle figure matplotlib.
ax = fig.add_subplot(111, projection='3d') # Ajoute un sous-graphe en 3D à la figure.
ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color='red', label='Données réelles')   # Affiche les données réelles sous forme de points rouges dans l'espace 3D.
plt.show() # Affiche le premier graph.
```

```python
model = LinearRegression()  # Crée une instance du modèle de régression linéaire.
model.fit(x_test, y_test)  # Entraîne le modèle avec les variables d'entrée x_test et la cible y_test.
y_prediction = model.predict(x_test)  # Calcule les prédictions du modèle à partir des données x_test.

print("Coefficients:", model.coef_)  # Affiche les coefficients appris par le modèle pour chaque variable.
print("Intercept:", model.intercept_)  # Affiche la constante du modèle, c'est-à-dire l'ordonnée à l'origine.

fig = plt.figure() # Crée une nouvelle figure pour la visualisation du modèle.
ax = fig.add_subplot(111, projection='3d') # Ajoute un sous-graphe 3D à cette nouvelle figure

ax.scatter(x_test[:, 0], x_test[:, 1], y_test, color='red', label='Données réelles')# Affiche à nouveau les données réelles sous forme de points rouges.
ax.plot_trisurf(x_test[:, 0], x_test[:, 1], y_prediction, color='blue', alpha=0.5, label='Plan de régression')# Trace une surface triangulée bleue représentant le plan de régression prédit.

ax.set_xlabel('X') # Définit le label de l'axe X.
ax.set_ylabel('Y') # Définit le label de l'axe Y.
ax.set_zlabel('Target') # Définit le label de l'axe Z.

ax.legend() # Affiche la légende du graphique.
plt.show() # Affiche le graphique final.
```
