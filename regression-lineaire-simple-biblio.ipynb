```
import numpy as np  # Importe la bibliothèque NumPy et lui donne l'alias np pour manipuler des tableaux et faire des calculs numériques.
from sklearn.linear_model import LinearRegression  # Importe la classe LinearRegression de scikit-learn pour créer un modèle de régression linéaire.
from sklearn.datasets import make_regression  # Importe la fonction make_regression pour générer un jeu de données de régression artificiel.
import matplotlib.pyplot as plt  # Importe le module pyplot de matplotlib sous le nom plt pour créer des graphiques.
```

```
np.random.seed(0)  # Fixe la graine aléatoire à 0 pour obtenir les mêmes données à chaque exécution.
x_test, y_test = make_regression(n_samples=200, n_features=1, noise=10)  # Génère 200 échantillons de données avec 1 variable explicative et un bruit de 10.
plt.scatter(x_test, y_test, color='green')  # Affiche les données générées sous forme de nuage de points verts.
```

```
model = LinearRegression()  # Crée une instance vide du modèle de régression linéaire.
model.fit(x_test, y_test)  # Entraîne le modèle en utilisant les données d'entrée x_test et les valeurs cibles y_test.
a = model.coef_[0]  # Récupère le coefficient directeur de la droite de régression appris par le modèle.
b = model.intercept_  # Récupère l'ordonnée à l'origine de la droite de régression.
print(a)  # Affiche la valeur du coefficient directeur.
print(b)  # Affiche la valeur de l'ordonnée à l'origine.

y_predict = model.predict(x_test)  # Utilise le modèle entraîné pour prédire les valeurs de sortie à partir de x_test.
plt.scatter(x_test, y_test, color='green', label='Data')  # Réaffiche les données originales sous forme de points verts avec une étiquette pour la légende.
plt.plot(x_test, y_predict, label='Ligne de regression')  # Trace la droite de régression correspondant aux prédictions du modèle.
plt.legend()  # Affiche la légende du graphique.
plt.show()  # Affiche le graph final à l'écran.
```
