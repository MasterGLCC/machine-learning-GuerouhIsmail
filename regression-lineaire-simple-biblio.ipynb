```
#On importe les biblio qu'on va utiliser
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```
#On génére notre data
np.random.seed(0)
x_test, y_test = make_regression(n_samples=200, n_features=1, noise=10)
plt.scatter(x_test, y_test, color='green')
```

```
#On crée un objet de régression linéaire vide
model = LinearRegression()
#On entraîne le modèle avec nos données X_test et Y_test
model.fit(x_test, y_test)
a = model.coef_[0]
b = model.intercept_
print(a)
print(b)
#On dessine notre courbe
y_predict = model.predict(x_test)
plt.scatter(x_test, y_test, color='green', label='Data')
plt.plot(x_test, y_predict, label='Ligne de regression')
plt.legend()
plt.show()
```
