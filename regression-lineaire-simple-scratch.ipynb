```
import numpy as np
```

```
#je définie la fonction regression_lineaire que je vais utiliser qui permet de prédire des valeurs inconnues en traçant une droite du meilleur ajustement
#les paramètres: X:valeur indépendante, y:valeur dépendante, a:la pente, b:l'odonnée à l'origine, alpha: valeur du pas, iteration: nombre d'itération
def regression_lineaire(X, y, a, b, alpha, iteration):
    #initialiser les variables m qui va prendre la longueur de la liste y et calcule_cout qui va stocker une liste des valeurs de coût au fil du temps
    m=len(y)
    calcule_cout=[]
    #loop
    for i in range (iteration):
        #calculer les predictions
        prediction=a*X + b
        #calculer le coût
        cout=((1/(2*m))*sum(prediction - y )**2)
        calcule_cout.append(cout)
        #calculer les gradients
        da=(1/m)*np.sum((prediction - y)* X)
        db=(1/m)*np.sum(prediction - y)
        #mettre à jour les nouvelles valeurs des paramètres a et b
        a=a-alpha*da
        b=b-alpha*db
    return a, b, calcule_cout
```
