```
import numpy as np
import matplotlib.pyplot as plt
```

```
def regression_lineaire(X, y, a, b, alpha, iteration):  # Définit la fonction de régression linéaire entraînée par descente de gradient.
    m = len(y)  # Récupère le nombre total d'exemples dans le vecteur y.
    calcule_cout = []  # Initialise une liste vide pour stocker l'évolution du coût.

    for i in range(iteration):  # Répète l'apprentissage pendant un certain nombre d'itérations.
        prediction = a * X + b  # Calcule les prédictions du modèle pour chaque valeur de X.
        cout = (1 / (2 * m)) * np.sum((prediction - y) ** 2)  # Calcule l'erreur quadratique moyenne divisée par 2.
        calcule_cout.append(cout)  # Ajoute la valeur actuelle du coût à la liste.
        da = (1 / m) * np.sum((prediction - y) * X)  # Calcule le gradient par rapport à la pente a.
        db = (1 / m) * np.sum(prediction - y)  # Calcule le gradient par rapport à l'intercept b.
        a = a - alpha * da  # Met à jour la pente a.
        b = b - alpha * db  # Met à jour l'intercept b.

    return a, b, calcule_cout  # Retourne les paramètres appris et la liste des coûts.

np.random.seed(0)  # Fixe la graine aléatoire pour obtenir les mêmes résultats à chaque exécution.
X = 2 * np.random.rand(100)  # Génère 100 valeurs aléatoires pour X entre 0 et 2.
y = 4 + 3 * X + np.random.randn(100)  # Génère y selon une relation linéaire bruitée.

a_initial = 0  # Initialise la pente à 0.
b_initial = 0  # Initialise l'ordonnée à l'origine à 0.
alpha = 0.1  # Définit le taux d'apprentissage.
iteration = 100  # Définit le nombre d'itérations de la descente de gradient.

a_final, b_final, calcule_cout = regression_lineaire(X, y, a_initial, b_initial, alpha, iteration)  # Entraîne le modèle et récupère les résultats.

print("Pente a :", a_final)  # Affiche la valeur finale de la pente.
print("Intercept b :", b_final)  # Affiche la valeur finale de l'ordonnée à l'origine.

y_pred = a_final * X + b_final  # Calcule les prédictions finales du modèle avec les paramètres appris.

plt.scatter(X, y, color='blue', label='Données réelles')  # Affiche les données réelles sous forme de points bleus.
plt.plot(X, y_pred, color='red', label='Droite de régression')  # Trace la droite de régression en rouge.
plt.xlabel('X') # Ajoute le label de l'axe des abscisses.
plt.ylabel('y') # Ajoute le label de l'axe des ordonnées.
plt.title('Régression linéaire') # Ajoute un titre au graph.
plt.legend() # Affiche la légende.
plt.show() # Affiche le graph des données et de la droite.
```
